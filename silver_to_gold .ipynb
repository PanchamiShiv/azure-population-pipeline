```python
# Cell 1 - Read from Silver
storage_account_name = "populationpipelinesa"
storage_account_key = "YOUR_STORAGE_ACCOUNT_KEY_HERE"  # use your actual key

silver_path = "abfss://silver@populationpipelinesa.dfs.core.windows.net/cities_cleaned"

df_silver = (spark.read
    .option("fs.azure.account.key.populationpipelinesa.dfs.core.windows.net", storage_account_key)
    .parquet(silver_path))

print("Rows:", df_silver.count())
df_silver.show(5)
```

```python
# Cell 2 - Top 10 Most Populated Cities
from pyspark.sql.functions import col, rank
from pyspark.sql.window import Window

df_top10 = df_silver \
    .orderBy(col("population").desc()) \
    .limit(10) \
    .select("Rank", "City", "Country", "population")

print("🏆 Top 10 Most Populated Cities:")
df_top10.show()
```

```python
# Cell 3 - Save to Gold container
gold_path = "abfss://gold@populationpipelinesa.dfs.core.windows.net/top10_populated_cities"

df_top10.write \
    .mode("overwrite") \
    .option("fs.azure.account.key.populationpipelinesa.dfs.core.windows.net", storage_account_key) \
    .parquet(gold_path)

print("✅ Gold layer saved successfully!")
```
